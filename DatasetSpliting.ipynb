```python
import pandas as pd
from sklearn.model_selection import train_test_split, StratifiedKFold

randomState = 42
```

---
#### Import the processed Data

```python
filename = ".\data\processed_data.csv"

data = pd.read_csv(filename)
display(data.head().iloc[:,:15])
display(data.head().iloc[:,15:])
display(data.describe().T)
```

```python
X = data.drop(columns=["DR"])
y = pd.DataFrame(data["DR"])

display(X.head().iloc[:,:18])
display(X.head().iloc[:,18:])
display(y.head())
```

---
#### First Split


```python
X_kfold, X_test, y_kfold, y_test = train_test_split(X, y, test_size=0.1, random_state=randomState,stratify=y)
```

---
#### Stratified K Fold

```python
kF = StratifiedKFold(n_splits=5,shuffle=True, random_state=randomState)

kFold = []

for fold, (train_idx, test_idx) in enumerate(kF.split(X_kfold,y_kfold)):
    train_x, train_y = X_kfold.iloc[train_idx], y_kfold.iloc[train_idx]
    test_x, test_y = X_kfold.iloc[test_idx], y_kfold.iloc[test_idx]

    kFold.append((train_x, test_x, train_y, test_y))
```

```python
import torch
from torch.utils.data import Dataset, DataLoader
import numpy as np

class CustomDataset(Dataset):
    def __init__(self, data, labels):
        assert len(data)==len(labels), "length of DATA and LABELS do not match"
        self.data = np.array(data)
        self.labels = np.array(labels)

    def __len__(self):
        # Return the size of the dataset
        return len(self.data)

    def __getitem__(self, idx):
        # Return a single data point and its corresponding label
        sample = torch.tensor(self.data[idx], dtype=torch.float32)
        label = torch.tensor(self.labels[idx], dtype=torch.float32)  # For classification tasks
        return sample, label
    

    
```

```python
from modularModels1 import modularNN , BlockMaker
# from modelV1 import model_1
import torch
import torch.nn as nn
import torch.optim as optim
import torch.nn.init as init
```

```python
model_1 = modularNN(1,
                   blocks=[BlockMaker(28,[128,128],64,
                                      [nn.ReLU(),nn.ReLU(),nn.LeakyReLU()]),
                           BlockMaker(64,[64],32,
                                      [(nn.ELU(),nn.Dropout(0.2)),nn.Sigmoid()]),
                           ],
                   blockSize=(28,32))

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
# print(device)
model_1 = model_1.to(device)
print("A", model_1)


for layer in model_1.modules():
    if isinstance(layer, (nn.Linear, nn.Conv2d)):  # Check if the layer has weights
        init.xavier_normal_(layer.weight)  # Initialize weights using Xavier Normal
        if layer.bias is not None:
            init.normal_(layer.bias, mean=0, std=0.01)  






# for name, params in model_1.named_parameters():
#     print(name, params)
```

```python
epochs = 5
batchSize = 64
criterion = nn.BCEWithLogitsLoss()
optimiser = optim.Adam(model_1.parameters(),lr=0.001)


# for fold, (train_x, test_x, train_y, test_y) in enumerate(kFold):

train_dataset = CustomDataset(X_kfold, y_kfold)
test_dataset = CustomDataset(X_test, y_test)

print(len(train_dataset))

train_loader = DataLoader(train_dataset, batch_size=batchSize, shuffle=True)
test_loader = DataLoader(test_dataset, batch_size=batchSize, shuffle=False)

model_1.train()
print(f"Fold: {fold+1}")
for epoch in range(epochs):
    running_loss = 0.0
    correct_train = 0
    total_train = 0

    for batch, (inputs, labels) in enumerate(train_loader):
        inputs, labels = inputs.to(device), labels.to(device).squeeze()
        optimiser.zero_grad()
        outputs = model_1(inputs).squeeze()
        # print(outputs)
        # print("Labels",labels)
        loss = criterion(outputs, labels)
        loss.backward()
        optimiser.step()

        running_loss += loss.item()
        predicted = (torch.sigmoid(outputs) > 0.5)
        correct_train += (predicted == labels).sum().item()
        total_train += labels.size(0)

        print(f"  Batch: {batch+1}".ljust(10), 
                f"Loss: {running_loss/batchSize/(batch+1)}".ljust(12),
                f"Accuracy: {((correct_train/total_train)*100):.3f}%", end="\r")

    train_accuracy = 100 * correct_train / total_train
    avg_loss = running_loss / len(train_loader)

    print(f"- Epoch [{epoch+1}/{epochs}]".ljust(10),
            f"Loss: {avg_loss:.4f}".ljust(12),
            f"Accuracy: {train_accuracy:.2f}%".ljust(30))



model_1.eval()
correct = 0
total = 0
with torch.no_grad():
    for inputs, labels in test_loader:
        inputs, labels = inputs.to(device), labels.float().to(device).squeeze()
        outputs = model_1(inputs).squeeze()
        predicted = (torch.sigmoid(outputs) > 0.5).float()
        total += labels.size(0)
        correct += (predicted == labels).sum().item()

val_accuracy = (correct / total)*100
print(f"val acc: {val_accuracy:.2f}%\n")
```
